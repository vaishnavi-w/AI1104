```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
p=int(input("Enter the dimension "))
K=int(input("Enter the number of clusters "))

#number of points in each cluster
NUM_PTS = 2000

#generating K random means for the data points
#array of size (K, p) of ints between -0.5*K and 0.5*K is generated
A = np.random.randint(-0.5*K, 0.5*K, size=(K, p))
#covariance matrix (0.1 variance)
cov = 0.1*np.identity(p)

#generating clusters from a p-dimensional Gaussian distribution (means as rows of array A, 0.1 variance)
#generating first cluster
Y = np.random.multivariate_normal(A[0], cov, NUM_PTS).T
#generating the remaining clusters and appending them to the initial cluster points
for i in range(1, K):
  Y = np.append(Y, np.random.multivariate_normal(A[i], cov, NUM_PTS).T, axis = 1)

#generate K - centroids from a p-dimensional gaussian distribution (zero mean, K variance)
D = np.random.multivariate_normal(np.zeros(p), K*np.identity(p), K).T 

# Set stopping condition 
epsilon = 0.001
# Initialize error to a large value
error = 1.0

count = 0
while (error > epsilon):
    #number of points in each cluster is initialized to zero
    num = np.zeros(K)
    
    #clusters assigned to each point is initialized to zero
    cluster = np.zeros(K*NUM_PTS)
    
    #centroids are initialized to zero
    X = np.zeros((p,K))
    
    #update clusters based on distance
    for idx in range(K*NUM_PTS):
        cur_y = Y[:,idx]
        
        #distance of the point from all centroids is initialized to zero
        d = np.zeros(K)
        
        #calculating the distance of the point from all centroids
        for i in range(K):
          d[i] = np.linalg.norm(cur_y - D[:,i])
        
        #array is converted to a list 
        distance_vector = d.tolist()
        
        #finding the closest centroid
        min_idx = distance_vector.index(np.min(distance_vector))
        
        #updating centroid and count
        X[:,min_idx] += cur_y
        num[min_idx] += 1

        #updating cluster to the point
        cluster[idx] = min_idx + 1
    
    #centroids calculation
    #updating only those centroids that are assigned some data points
    for i in range(K):
        if num[i] > 0:
            X[:,i] = X[:,i]/num[i]
    
    #error calculation by computing distance between the old and updated centroids
    error = 0
    for i in range(K):
        error = error + np.linalg.norm(X[:,i] - D[:,i])
    
    count +=1
    
    #update centroids
    D=X

    # Print error
    print ('Iteration: ', count, 'Error: ', error)

#appending data and cluster labels
Y = np.append(Y, [cluster], axis=0)
#Y[p][i] gives the cluster label of the ith point 

# Print centroids
print ('Centroids:', D)
print ('Original means of clusters:', A.T)
```
